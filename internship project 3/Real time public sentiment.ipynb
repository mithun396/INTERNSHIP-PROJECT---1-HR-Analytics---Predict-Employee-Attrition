```python
import pandas as pd
import time
```

```python
cols = ["sentiment","id","date","query","user","tweet"]

df = pd.read_csv(
    "training.1600000.processed.noemoticon.csv",
    encoding="latin-1",
    names=cols
)
```

```python
print(df.head())
```

```python
df["sentiment"] = df["sentiment"].replace({0:"Negative",4:"Positive"})
```

```python
df = df.sample(5000)
```

```python
df.to_csv("tweets_sample.csv", index=False)
```
